```python
import random

name = input("Plese enter your name: ")
print(f"Hello {name}, welcome to this game of Hangman!")
print("You will recieve a blanked out word and 10 chances to guess the correct word.\nBest of Luck!")

# The words list
words = ["abruptly","absurd","abyss","affix","askew","avenue",'awkward','axiom','azure','bagpipes',
         'bandwagon','banjo','beans','beekeeper','blitz','blizzard','boggle','bookworm','boxcar',
         'boxful','butterfly','buffalo','buffoon','blizzard','buzzing','buzzwords','caliph','cobweb',
         'cockiness','croquet','crypt','cycle','disavow','dizzying','duplex','dwarves','embezzle',
         'equip','espionage','forest','fruits','greatness', 'hopeful','grave','wizard','wristwatch',
         'xylophone','yacht','yolk','youthful','yummy','zephyr','zigzag','zipper','zodiac','zombie']
# Initialise Game
word = words[random.randint(0, len(words)-1)]
if len(word) <= 5: # give one letter hint for small (<=5) string
    k = 1
else: # ELse give a 2 letter string
    k = 2
hint_index = random.sample(range(0, len(word)), k)
guesses = '' # Creatiing empty guess string
for hint in hint_index:
    guesses += word[hint]
chances = 10 # Initialising number of chances
play = 'YES' # Willingness to play

def playAgain():
    global play, hint_index, guesses, word, words, chances
    play = input("Do you wanna play again?(YES/NO): ")
    play = play.upper()
    if play == 'Y':
        play = 'YES'
    if play == 'YES':
        word = words[random.randint(0, len(words)-1)]
        if len(word) <= 5: # give one letter hint for small (<=5) string
            k = 1
        else: # ELse give a 2 letter string
            k = 2
        hint_index = random.sample(range(0, len(word)), k)
        guesses = '' # Creatiing empty guess string
        for hint in hint_index:
            guesses += word[hint]
        chances = 10 # Initialising number of chances
    else:
        print(f"Thank you for playing!")

while play == 'YES':
    
    while chances > 0:
        won = True # Changing win to True at beginning of every round
        for letter in word:
            if letter in guesses:
                print(letter, end = ' ')
            else:
                print('_', end = ' ')
                won = False # If even 1 _ is there winning is not possible
        if won == True:
            print(f"\nCongratulations! You Won! The word was: {word}")
            print(f"Your score is {chances*10}!")
            playAgain()
            break 
            # This only breaks from the 'chances' loop and and when play is 'YES', game continues
        guess = input("\nGuess your next letter: ")
        guess = guess.lower()
        if guess not in guesses:
            guesses += guess # Adding guess to guesses
            if guess not in word:
                chances -= 1
                print(f"{guess} is not in the word, you have {chances} chance(s) left")
                if chances == 0:
                    print(f"You Lose! The word was {word}. Better luck next time!")
                    playAgain()
                    break 
                    # This only breaks from the 'chances' loop and and when play is 'YES', 
                    # game continues

        else: 
            print(f"You have already guessed {guess}, guess a different letter.")
```
